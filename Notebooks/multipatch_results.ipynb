# Loading data and plot parameters

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
# Coarse
coarse_data_central = np.loadtxt("multipatch_results/Test_Multipatch_coarse_data_central.dat")
coarse_data_right = np.loadtxt("multipatch_results/Test_Multipatch_coarse_data_right.dat")

coarse_error_central = np.loadtxt("multipatch_results/Test_Multipatch_coarse_error_central.dat")
coarse_error_right = np.loadtxt("multipatch_results/Test_Multipatch_coarse_error_right.dat")

# Medium
medium_data_central = np.loadtxt("multipatch_results/Test_Multipatch_medium_data_central.dat")
medium_data_right = np.loadtxt("multipatch_results/Test_Multipatch_medium_data_right.dat")

medium_error_central = np.loadtxt("multipatch_results/Test_Multipatch_medium_error_central.dat")
medium_error_right = np.loadtxt("multipatch_results/Test_Multipatch_medium_error_right.dat")

# Fine
fine_data_central = np.loadtxt("multipatch_results/Test_Multipatch_fine_data_central.dat")
fine_data_right = np.loadtxt("multipatch_results/Test_Multipatch_fine_data_right.dat")

fine_error_central = np.loadtxt("multipatch_results/Test_Multipatch_fine_error_central.dat")
fine_error_right = np.loadtxt("multipatch_results/Test_Multipatch_fine_error_right.dat")

# Convergence
convergence_central = np.loadtxt("multipatch_results/Test_Multipatch_convergence_center.dat")
convergence_right = np.loadtxt("multipatch_results/Test_Multipatch_convergence_right.dat")

# Plot aesthetics

font_size = 20

mpl.rcParams['mathtext.fontset'] = 'cm'
mpl.rcParams['font.family'] = 'Latin Modern Roman'
plt.rcParams['figure.figsize'] = [10, 8]
mpl.rcParams['xtick.labelsize'] = font_size
mpl.rcParams['ytick.labelsize'] = font_size
```

# Coarse data results

```python
plt.close('all')
plt.plot(coarse_data_central[:,0], coarse_data_central[:,1])
plt.plot(coarse_data_right[:,0], coarse_data_right[:,1])

plt.title('Central and +X patch data', fontsize = font_size);
plt.xlabel(r'$x$', fontsize = font_size);
plt.ylabel('K_Phi_RHS', fontsize = font_size);
```

```python
plt.close('all')
plt.plot(coarse_error_central[:,0], coarse_error_central[:,1])

plt.title('Central patch error', fontsize = font_size);
plt.xlabel(r'$x$', fontsize = font_size);
plt.ylabel('K_Phi_RHS', fontsize = font_size);
```

```python
plt.close('all')
plt.plot(coarse_error_right[:,0], coarse_error_right[:,1])

plt.title('+X patch error', fontsize = font_size);
plt.xlabel(r'$x$', fontsize = font_size);
plt.ylabel('K_Phi_RHS', fontsize = font_size);
```

# Convergence

```python
plt.close('all')
plt.scatter(convergence_central[:,0], convergence_central[:,1])

plt.title("Central convergence", fontsize = font_size);
plt.xlabel(r'$x$', fontsize = font_size);
plt.ylabel('K_Phi_RHS', fontsize = font_size);
plt.ticklabel_format(useOffset=False, style='plain')
```
